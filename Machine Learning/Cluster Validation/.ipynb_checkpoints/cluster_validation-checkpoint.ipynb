```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.DataFrame({
    'x':[0.2,0.6,0.3,0.7],
    'y':[0.9,0.3,0.8,0.2]
})
data
```

```python
plt.scatter(data['x'],data['y']);
```

```python
#use K-means clusturing algorithm 
#train
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2)
model.fit(data)
label = model.predict(data)
```

```python
#predictions
label
```

```python
#getting the clusters using different colors according to label
plt.scatter(data['x'], data['y'], c=label);
```

```python
#calculate the Silhouette Score
from sklearn.metrics import silhouette_score
sc = silhouette_score(data, label, metric = 'euclidean')
print(f'Silhouette Score:{sc}')
```
